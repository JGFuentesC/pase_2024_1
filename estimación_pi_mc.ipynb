## Dependencias

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Generación de números aleatorios

```python
X = 2*np.random.uniform(size=100)-1
Y = 2*np.random.uniform(size=100)-1
```

```python
plt.scatter(X,Y)
```

## Estimación de $\pi$

```python
def I(x:float,y:float)->float:
    return 1 if x**2+y**2<=1 else 0
```

```python
4*np.vectorize(I)(X,Y).mean()
```

## Simulación 

```python
l = []
for k in range(1000,100000,1000):
    X = 2*np.random.uniform(size=k)-1
    Y = 2*np.random.uniform(size=k)-1
    l.append((k,4*np.vectorize(I)(X,Y).mean()))
sim = pd.DataFrame(l,columns=['k','pi']).set_index('k')
sim['baseline'] = np.pi
sim.plot(kind='line')
```
