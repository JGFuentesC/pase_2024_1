## Dependencias

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Generación a través del método de aceptación-rechazo

```python
def genNorm():
    while True:
        Y1,Y2 = np.random.exponential(scale=1),np.random.exponential(scale=1)
        C = Y2-(Y1-1)**2/2
        if np.sign(C)>0:
            Y1 = C
            U = np.random.uniform()
            Z = Y1 if U<=1/2 else -Y1
            return Z
            break 
```

```python
norm = np.array([genNorm() for i in range(10000)])
```

## Generación de normal estándar numpy 

```python
norm_t = np.random.normal(size=10000)   
```

## Comparación 

```python
sns.kdeplot(norm, label='Aceptación-Rechazo')
sns.kdeplot(norm_t, label='Numpy')
```
