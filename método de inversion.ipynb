## Dependencias

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Generación de números uniformes 

```python
U = np.random.uniform(size=10000)
```

## Generación de números exponenciales 

### Método de la transformada inversa

```python
exp = -np.log(U)
```

### Números exponenciales $\lambda=1$ a través de otro método (el estándar de numpy)

```python
exp_t = np.random.exponential(size=10000, scale=1)  
```

## Comparación de histogramas

```python
sns.kdeplot(exp, label='Método de la transformada inversa')
sns.kdeplot(exp_t, label='Método de numpy')
```
